```python
import pandas as pd
df = pd.read_csv('/content/sample_dataset_100.csv', names=["Name", "Category", "Price (PKR)", "Rating","product_description"])
df = df.drop(index=0).reset_index(drop=True)
print(df)
```

```python
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem import SnowballStemmer
ps= SnowballStemmer('english')
```

```python
from nltk.tokenize import word_tokenize
nltk.download('punkt_tab')
corpus=[]
for i in range (0, len(df)):
  review= re.sub('[^a-zA-Z]', ' ', df['product_description'][i])
  review= review.lower()
  review= word_tokenize(review)
  review=[ps.stem(word) for word in review if word not in set(stopwords.words('english'))]
  review= ' '.join(review)
  corpus.append(review)

print(corpus)


```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv= CountVectorizer(max_features= 20, binary=True) #maximum number of frequency
x=cv.fit_transform(corpus).toarray()
x.shape
x
```
