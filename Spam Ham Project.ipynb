```python
import pandas as pd
url = "https://raw.githubusercontent.com/krishnaik06/Complete-Data-Science-With-Machine-Learning-And-NLP-2024/refs/heads/main/26-CompleteNLP%20For%20Machine%20Learning/Practicals/SpamClassifier-master/smsspamcollection/SMSSpamCollection"
df = pd.read_csv(url, sep='\t', names=['label', 'messages'])
display(df.head())
```

```python
  import nltk
  nltk.download('stopwords')
  from nltk.corpus import stopwords
  from nltk.stem import PorterStemmer
  import re
  ps = PorterStemmer()
  nltk.download('punkt_tab')
  from nltk.tokenize import word_tokenize, sent_tokenize
  corpus=[]
  df['token_sentence']= df['messages'].apply(sent_tokenize)
  df['word_tokenize']= df['messages'].apply(word_tokenize)
  df
```

```python
for i in range(0, len(df)):
  review= re.sub('[^a-zA-Z]', ' ', df['messages'][i])
  review= review.lower()
  review= word_tokenize(review)
  review=[ps.stem(word) for word in review if word not in set(stopwords.words('english'))]
  review= ' '.join(review)
  corpus.append(review)
corpus
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv= CountVectorizer(max_features=100, ngram_range=(1,2))
X= cv.fit_transform(corpus).toarray()
X
```

```python
cv.vocabulary_
```

```python
print(X.shape)       # should be (num_docs, 100)
print(X[0])          # vector for first document
print(X[0].nonzero())# indices of non-zero features
```

y= pd.get_dummies(df['label'])

This is because ML models need numbers, not text labels. "y" becomes a DataFrame with two columns.

```python
#Output Feature & Dependent Feature
y= pd.get_dummies(df['label'])
y
```

**Explanation**:

Imports "train_test_split" from sklearn. This function splits data into training and testing sets to evaluate how well our model learns without cheating (seeing the test data during training).

Splits "X" (features) and "y" (labels) into four parts: training features (X_train), testing features (X_test), training labels (y_train), and testing labels (y_test). "test_size=0.20" means 20% for testing, 80% for training. It shuffles randomly for fairness. This is like dividing study materials: train on most, test on the rest to check understanding.

```python
#Train Test Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X,y, test_size=0.20)
```

**What is Naive Bayes?**

Naive Bayes is a simple ML method that uses probabilities (chances) to make predictions. It’s "naive" because it assumes things are simpler than they might be—like assuming each word in a message contributes independently to whether it’s spam or not, ignoring how words might relate (e.g., "free gift" together might mean more than "free" alone).

**Why Multinomial?**

The "multinomial" part means it’s good for counting things, like how many times words appear in messages. In our case, we’re counting words (from CountVectorizer) to figure out if a message is spam or ham.

**Why use it?**

It’s easy to understand, fast, and works great for text problems like spam detection, where certain words (e.g., "free," "win") are strong clues for spam. It’s like a beginner-friendly tool that still gets good results.

The dog learns patterns (e.g., strangers wear hats, family doesn’t). Here, X_train is the examples (messages), and y_train['spam'] is the instructions (spam or not).

```python
from sklearn.metrics import accuracy_score,classification_report
```

**Code**:

***accuracy_score(y_test['spam'], y_pred)***

**Simple Explanation**:

This compares the model’s predictions (y_pred) to the true labels (y_test['spam']) and calculates what percentage it got right.
Example: If there are 100 test messages and the model correctly predicts 95 as spam or ham, the accuracy is 95% (0.95).
Analogy: It’s like grading a multiple-choice test. If the student (model) gets 95 out of 100 questions right, their score is 95%.


**Deeper Insight for Beginners**:

**How it works**:

accuracy_score counts how many times y_pred matches y_test['spam'] and divides by the total number of test messages. Formula: (correct predictions) / (total predictions).

**Example**:

If y_test['spam'] = [0, 1, 0, 0, 1] and y_pred = [0, 1, 0, 1, 1], it got 4/5 correct (mismatched on the 4th), so accuracy is 4/5 = 0.8 (80%).
For your data: The SMS Spam Collection dataset has ~5572 messages, so 20% (~1115) are in X_test. Accuracy is typically high (95–98%) because Naive Bayes is good at spotting spam patterns (e.g., words like "free" or "win").

**Limitation**:

Accuracy can be misleading if the data is imbalanced (e.g., ~87% ham, 13% spam in this dataset). A model predicting "ham" always could get ~87% accuracy but fail at catching spam. That’s why you added classification_report.

```python
accuracy_score(y_test['spam'],y_pred)
```

**Simple Explanation**
This creates a detailed report showing how well the model did at predicting spam (1) and ham (0). It gives three scores for each: precision, recall, and F1-score, plus an overall summary.

**Deeper Insight for Beginners**:

What’s in the report? The classification_report provides:


**Precision**: How often the model is right when it predicts a class. For spam: (correct spam predictions) / (total spam predictions). Example: If it predicts 100 messages as spam and 90 are actually spam, precision is 90/100 = 0.90 (90%). High precision means few false alarms (e.g., not marking your friend’s email as spam).

**Recall**: How many of the actual class instances the model caught. For spam: (correct spam predictions) / (total actual spam messages). Example: If there are 100 spam messages and it catches 85, recall is 85/100 = 0.85 (85%). High recall means it misses few spams.
F1-score: A balance between precision and recall. It’s the harmonic mean: 2 * (precision * recall) / (precision + recall). If precision is 0.90 and recall is 0.85, F1 ≈ 0.87. It’s useful when you want both to be good.

**Support**: The number of actual instances of each class in y_test['spam'] (e.g., 965 ham, 150 spam in a typical test split).
Macro avg: Average of precision, recall, and F1 across classes (treats spam and ham equally).

**Weighted avg**: Same as macro avg but weighted by support (accounts for more ham than spam).

```python
from sklearn.metrics import classification_report
print(classification_report(y_test['spam'],y_pred))
```

These results come from a machine learning model designed to classify text messages as either **ham** (legitimate messages) or **spam**. Here’s a breakdown of what the report means in plain English:

---

### Understanding the Results

* **Ham (0):** The model is excellent at identifying ham messages.
    * **Precision (98%):** When the model says a message is ham, it's correct 98% of the time.
    * **Recall (99%):** It successfully identifies 99% of all the actual ham messages.
    * **F1-score (0.98):** This score shows the model's high overall performance for ham messages.

* **Spam (1):** The model is a little less effective at catching spam.
    * **Precision (92%):** When the model says a message is spam, it's correct 92% of the time.
    * **Recall (85%):** It misses 15% of all the actual spam messages, letting some junk through.
    * **F1-score (0.88):** While still good, this score is lower than for ham, indicating a weaker performance.

### Key Takeaways

* **Accuracy (97%):** The model is 97% accurate overall. This sounds great, but it’s a bit misleading because most of the messages were ham.
* **Why Spam Performance is Lower:** The model had far fewer spam messages to learn from (150 spam messages vs. 965 ham messages). With less data, it was harder for the model to learn and recognize the patterns of spam, which is why its **recall** is lower.
* **The Big Picture:** This report shows the model is great at classifying legitimate messages, but it’s not as good at catching spam. In a real-world spam filter, you'd probably want to improve the **spam recall** to catch as much junk as possible, even if it means a few legitimate messages are incorrectly marked as spam.

This is a perfect example of why you should look at more than just a model's overall accuracy. A high accuracy score can hide significant problems, like a spam filter that misses a lot of spam.
