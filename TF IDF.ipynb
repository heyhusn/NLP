```python
import pandas as pd
df= pd.read_csv('/content/sample_5col_100rows.csv', names=["product_num", "category", "price pkr", "status", "product_description"])
df= df.drop(index=0).reset_index(drop=True)
df
```

```python
import nltk
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
lemma= WordNetLemmatizer()
```

```python
corpus=[]
for i in range(len(df)):
  review= re.sub('[^a-zA-Z]', ' ', df['product_description'][i])
  review= review.lower()
  review= word_tokenize(review)
  review= [lemma.lemmatize(word) for word in review if word not in set(stopwords.words('english'))]
  review= ' '.join(review)
  corpus.append(review)

corpus
```

```python
import sklearn
from sklearn.feature_extraction.text import TfidfVectorizer
vector= TfidfVectorizer(max_features=50)
vector.fit_transform(corpus).toarray()


```
